# removing outlier

```python
numerical_columns = ['age', 'weight', 'ball_controlling_skills', 'body_reflexes', 'jumping_skills', 'penalties_conversion_rate']
# Calculate the IQR for each column
Q1 = df[numerical_columns].quantile(0.25)
Q3 = df[numerical_columns].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers from the dataset
df_no_outliers = df[~((df[numerical_columns] < lower_bound) | (df[numerical_columns] > upper_bound)).any(axis=1)]

# Display the original and outlier-free datasets
print("Original dataset shape:", df.shape)
print("Dataset shape after removing outliers:", df_no_outliers.shape)
```

```python
df_no_outliers = df_no_outliers.drop(['name', 'player_id'], axis=1)

# Display the updated dataset
print(df_no_outliers)
```

```python
# Convert height to centimeters
df_no_outliers['height'] = df_no_outliers['height'].apply(lambda x: int(x.split("'")[0]) * 30.48 + int(x.split("'")[1]) * 2.54)

# Display the updated dataset
print(df_no_outliers)
```
